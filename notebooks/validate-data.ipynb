```python
import os
from glob import glob

import matplotlib.pyplot as plt
from tqdm import tqdm

from skimage.io import imread
from skimage.color import label2rgb
```

```python
medical_data_path = "/home/akamath/Documents/" + "toskipornot/data/Task02_Heart-processed/"

fold = "train"
medical_data_images = os.path.join(medical_data_path, fold, "images")
image_data = sorted(glob(os.path.join(medical_data_images, "*.png")))

medical_data_labels = os.path.join(medical_data_path, fold, "labels")
label_data = sorted(glob(os.path.join(medical_data_labels, "*.png")))
```

```python
n_images = 20 #len(images)

for image_index in tqdm(range(n_images)):
    image = imread(image_data[image_index])
    label = imread(label_data[image_index])
    overlay = label2rgb(label, image=image, bg_label=0, alpha=0.25)

    # plot representative images
    plt.figure()
    fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(9, 6))
    plt.gray()
    ax[0].imshow(image)
    ax[1].imshow(label)
    ax[2].imshow(overlay)
    plt.title("Data point: " + str(image_index))
```
